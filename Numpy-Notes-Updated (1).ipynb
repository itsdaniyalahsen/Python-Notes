### Slicing (Very Important)

The python 'comma' "**,**" says go through all, For instance when I apply this to rows.

1) $Array[:,i]$ Fetches the $i+1$th element from each column

2) $Array[i:, :j]$ Fetches the $0$th through $j$th, $j$th element exclusive from an array. Starting from the $i$th row to the end.

3) $Array[i:, j:k]$ Fetch the $j$ through $k$ elements in the Array with the $k$th element begin excluded.

4) $Array[i:j,l:k]$ Fetches the $l$ through $k$th elements ($k$th element excluded) in $i$ through $j$ rows in the array.

Notice  that arrays aren't indexed like $Array[i:j][j:k]$, This would cause problems. The comma replaces the $][$ square brackets in the middle

```python
a1 = np.random.randn(2,3)
a1
```

```python
# Demonstration of (1)
a1[:,0] # Go across all rows and fetch me the first column
a1[:,1] # Fetch the first element
```

```python
# Demonstration of (2)
a1[1:, :2]
```

```python
# Demonstration of (3)
a1[1:, 1:2]
```

```python
# Demonstraion of (4)
a = np.random.random((10,10))
a[8:, :2]
```

### Exercise 5:

Find a solution to $(\frac{d^2}{dx^2} + (10x)^2)f = \lambda f$ with boundary conditions $f(0) = f(1) = 0$

**Hint**: Use $numpy.diag( diagonal  , n )$ ,

where n is how many rows do yo go up from the principal diagonal, negative if you want to go below

```python
# First specify the No. of points
N = 100
x = np.linspace(0,1, N+1)
dx = np.diff(x)[0]
```

```python
main_diag = -2 * np.ones(N-1)
off_diag1, off_diag2 = np.ones(N-2),np.ones(N-2)

derivative_matrix = np.diag(main_diag) + np.diag(off_diag1, 1) + np.diag(np.diag(off_diag2, -1))
derivative_matrix /= dx**2
operator_10_x_sq_matrix = np.diag(10*x[1:-1]**2)

LHS_matrix = derivative_matrix + operator_10_x_sq_matrix
```

```python
eigen_val, eigen_vec = np.linalg.eigh(LHS_matrix) # eigh - This calculates eigenvalues for symmetric matrices
```

```python
pp.plot(eigen_vec[:, 0])
pp.plot(eigen_vec[:, 1])
pp.plot(eigen_vec[:, 2])
```

### Reading External File Data:

Use numpy.$loadtxt$ with $dtype$ = $object$. 

For CSV files, $delimiter = ','$

And for unpacking, depending on the number of columns use,

$Col_1$, $Col_2$, ..., $Col_n$ = $np.loadtxt( Location, dtype='object', delimiter=',', unpack=True )$

Finally to skip any specified number of rows, use

$skiprows = n$

For n rows,

In particular to skip the first row (which contains headings and is mostly useless), use

$Col_1$, $Col_2$, ..., $Col_n$ = $np.loadtxt( Location, dtype='object', delimiter=',', unpack=True, skiprows=1 )$

Now there is only one catch, All objects in $Col_i$ will have type 'object', to fix this, use

$Col_i = Col_i.astype(type)$
